```python
import matplotlib.pyplot as plt
import math


n = 100

g = lambda x: math.sin(x+math.sin(x))
xvec = [10*(i/n) for i in range(n)]

def derivative(func, x, i, h):
    return (func(x[i]+h)-func(x[i]-h))/(2*h)

gdot = [derivative(g, xvec, i, 1e-4) for i in range(n-2)]

plt.plot(xvec, [g(xvec[i]) for i in range(len(xvec))])

plt.plot(xvec[:-2], gdot, label = 'Numerical Derivative')

#plt.plot(xvec, [math.cos(xvec[i]) for i in range(len(xvec))], label = 'Analytic Derivative')

plt.title('Numerical Derivative of $f(x) = \\sin(x)$ Using Central Difference')

plt.legend()
plt.show()
```
